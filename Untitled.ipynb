```python
import joblib
import numpy as np
import matplotlib.pyplot as plt

# Load feature list
features = joblib.load("models/features_list.joblib")

# Load full model
model = joblib.load("models/asd_model.joblib")

# Extract the RandomForest model from the VotingClassifier
rf = model.base_estimators_[0]   # assuming RF is first in the list

importances = rf.feature_importances_
indices = np.argsort(importances)[::-1]

plt.figure(figsize=(10, 6))
plt.bar(range(len(importances)), importances[indices])
plt.xticks(range(len(importances)), [features[i] for i in indices], rotation=45)
plt.title("Feature Importance (RandomForest)")
plt.tight_layout()
plt.show()
```
